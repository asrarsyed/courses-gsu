#### 1. Identify missing values [6 pts]

**Task.**  Some values in the dataset may not be properly formatted. Replace any suspicious placeholders with standard missing values, and generate a summary table showing the count and proportion of missing data for affected columns.

```python
# Replace '?' with np.nan
df.replace('?', np.nan, inplace=True)

# Count the missing values
missing_counts = df.isnull().sum()

# Proportion of missing values
missing_proportions = df.isnull().mean()

# Summary table
missing_summary = pd.DataFrame({
    'Missing Count': missing_counts,
    'Missing Proportion': missing_proportions
})

# Filter only columns with missing data
missing_summary = missing_summary[missing_summary['Missing Count'] > 0]

missing_summary
```

#### 2. Prepare the target variable [6 pts]

**Task.**  
Extract the column representing the outcome of interest.  
Check whether it contains any missing values:
- If missing values are present, remove the corresponding rows from the dataset.
- If there are no missing values, encode the target as a binary variable suitable for classification.
- Store the result in a variable called `y`.
- Store the remaining features in a variable called `X`.

```python
# Check for missing values in the target column
print(df['income'].isnull().sum())

# Encode the target variable as binary: '>50K' as 1, '<=50K' as 0
df['income_encoded'] = df['income'].map({'>50K': 1, '<=50K': 0})

# Check for any remaining NaNs after encoding
print(df['income_encoded'].isnull().sum())

# Check the unique values of the encoded 'income_encoded' column
print(df['income_encoded'].unique())  # This should print: [0, 1]

# Drop target from the feature set
X = df.drop('income', axis=1)  # Features (excluding the original 'income')

# Define target variable y as the encoded target column
y = df['income_encoded']  # Encoded target
```

#### 3. Add missing indicators [4 pts]

**Task.**  For each column with missing values, create a new binary indicator variable that records whether the value was missing (`1`) or not (`0`) for that observation.

```python
# Binary indicator columns for missing values
for column in df.columns:
    if df[column].isnull().any():
        df[f'{column}_missing'] = df[column].isnull().astype(int)

# View result
df.head()
```

```python
df.filter(like='_missing').sum()
```

#### 4. Impute with random sampling [6 pts]

**Task.**  For each column with missing values, fill the missing entries by randomly sampling from the non-missing values of the same column. Ensure the imputation is reproducible by setting a random seed.

```python
# Random seed for reproducibility
np.random.seed(42)

# Loop through each column with missing values
for column in df.columns:
    if df[column].isnull().any():
        # Sample from non-missing values and fill the missing entries
        non_missing_values = df[column].dropna()
        df[column] = df[column].apply(lambda x: np.random.choice(non_missing_values) if pd.isnull(x) else x)

# View result
df.isnull().sum()
```

#### 5. Detect outliers with IQR rule [6 pts]

**Task.**  Using the Interquartile Range (IQR) rule, identify outliers in each numerical column. Report the **proportion** of outliers per column (relative to the dataset size).

```python
# Select numerical columns
numerical_cols = df.select_dtypes(include='number').columns

# Store proportions in a dictionary
outlier_proportions = {}

# Total number of rows for proportion calculation
n_rows = df.shape[0]

# Detect outliers for each numerical column
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    proportion = len(outliers) / n_rows
    outlier_proportions[col] = proportion

# Convert to a DataFrame for viewing
outlier_df = pd.DataFrame.from_dict(outlier_proportions, orient='index', columns=['Outlier Proportion'])
outlier_df.sort_values(by='Outlier Proportion', ascending=False)
```

Standardize the numerical variables so they have zero mean and unit variance.

```python
# Identify original numeric columns (excluding indicators & binary features)
original_numerics = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the selected columns
df[original_numerics] = scaler.fit_transform(df[original_numerics])
```

#### 8. One-hot encode categorical variables [4 pts]

**Task.**  
Apply one-hot encoding to all categorical variables in the dataset.

- Use `pd.get_dummies()` with `drop_first=True` to avoid multicollinearity.

#### 9. Remove constant and quasi-constant features [4 pts]

**Task.**  
Identify and remove features that carry little to no variability.  
Specifically, drop any column in which a single value appears in more than 99.995% of the observations.

```python
# Threshold for quasi-constant features
threshold = 0.99995

# Find columns where the most common value proportion exceeds the threshold
quasi_constant_cols = [
    col for col in df.columns
    if df[col].value_counts(normalize=True).iloc[0] > threshold
]

# Drop those columns
df = df.drop(columns=quasi_constant_cols)

# Print removed columns for inspection
print(f"Removed {len(quasi_constant_cols)} quasi-constant feature(s): {quasi_constant_cols}")
```

#### 10. Remove highly correlated features [8 pts]

**Task.**  
1. Plot a heatmap of the absolute correlation matrix for all numerical features.  
   You may use `matplotlib.pyplot` to set the figure size (e.g., `plt.figure(figsize=(24, 24))`).
2. Identify pairs of features with strong linear correlation (absolute value of correlation > 0.70).
3. From each highly correlated pair, remove one feature to reduce redundancy. Report features that you drop.

```python
import matplotlib.pyplot as plt
```

```python
# Only use the standardized original numerics for correlation analysis
corr_matrix = df[original_numerics].corr().abs()

# Plot the heatmap
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Absolute Correlation Matrix of Numerical Features")
plt.show()

# Identify pairs with correlation > 0.70 (excluding self-correlations)
high_corr_pairs = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if corr_matrix.iloc[i, j] > 0.70:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            high_corr_pairs.add((col1, col2))

# Drop one feature from each high correlation pair (keep the first, drop the second for simplicity)
features_to_drop = {pair[1] for pair in high_corr_pairs}
df = df.drop(columns=features_to_drop)

# Report dropped
print(f"Dropped {len(features_to_drop)} highly correlated feature(s): {list(features_to_drop)}")
```

#### 11. Forward feature selection [12 pts]

**Task.**  
Perform step-forward feature selection using a logistic regression classifier with `max_iter=1000` and `random_state=42`.

- Use **F1 score** as the evaluation metric.
- Use 5-fold cross-validation.
- Select the top 4 features that yield the best cross-validated F1 score.
- Use `SequentialFeatureSelector` from `sklearn.feature_selection`.
- Be mindful of which portion of the data is used to fit the feature selector and where you apply it.
- Report selected features.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, f1_score
```

```python
# Define features (X) and target (y)
X = df.drop(columns=['income_encoded'])  # Features excluding the target column
y = df['income_encoded']  # Update target variable to 'income_encoded'
```

```python
# Initialize Logistic Regression model
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Define F1 scorer for evaluation
f1_scorer = make_scorer(f1_score)

# Perform step-forward feature selection with 5-fold cross-validation
selector = SequentialFeatureSelector(logreg, n_features_to_select=4, direction='forward', scoring=f1_scorer, cv=5)

# Fit the feature selector to the data (X, y)
selector.fit(X, y)

# Get the selected features (those that have been selected)
selected_features = X.columns[selector.get_support()].tolist()

# Print selected features
print("Selected Features:", selected_features)
```

- Use **F1 score** as the evaluation metric.
- Use 5-fold cross-validation.
- Select the top 4 features that yield the best cross-validated F1 score.

```python
# Initialize Logistic Regression model
logreg = LogisticRegression(max_iter=1000, random_state=42)

# Define F1 scorer for evaluation
f1_scorer = make_scorer(f1_score)

# Perform step-backward feature elimination with 5-fold cross-validation
selector_backward = SequentialFeatureSelector(
    logreg,
    n_features_to_select=4,
    direction='backward',
    scoring=f1_scorer,
    cv=5,
    n_jobs=-1  # Use all CPUs to speed things up
)

# Fit the feature selector to the data (X, y)
selector_backward.fit(X, y)

# Get the selected features
selected_features_backward = X.columns[selector_backward.get_support()].tolist()

# Print selected features
print("Selected Features (Backward Selection):", selected_features_backward)
```

#### 13. Recursive feature elimination [15 pts]

**Task.**  
Implement a recursive feature elimination procedure using a Decision Tree classifier.

- Rank features by importance using the fitted Decision Tree.
- Iteratively remove the least important feature and retrain the model on the remaining set.
- After each iteration, evaluate model performance using **F1 score**.
- If the performance drops by more than a small threshold = 0.01, consider the removed feature important and keep it.
- Continue until all features have been evaluated.

üí° You may choose to use a custom implementation or a loop-based approach ‚Äî **do NOT use `RFECV` or `RFE` from `sklearn`**.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Use a increasingly larger sample size
X_sample = X.sample(n=48842, random_state=42)
y_sample = y.loc[X_sample.index]

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=42)

# Initial model
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X_train, y_train)
baseline_f1 = f1_score(y_test, model.predict(X_test))

selected_features = []
threshold = 0.01  # F1 drop threshold

for feature in X_train.columns:
    # Drop one feature at a time
    X_train_reduced = X_train.drop(columns=[feature])
    X_test_reduced = X_test.drop(columns=[feature])
    
    model = DecisionTreeClassifier(max_depth=5, random_state=42)
    model.fit(X_train_reduced, y_train)
    reduced_f1 = f1_score(y_test, model.predict(X_test_reduced))
    
    # If F1 drops significantly, keep the feature
    if baseline_f1 - reduced_f1 > threshold:
        selected_features.append(feature)

print("Selected Features (RFE):", selected_features)
```
