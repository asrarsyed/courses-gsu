#### 7. What is `Unnamed: 0` column? Can we delete it without losing essential information? Why?

The Unnamed: 0 column usually appears when a CSV or data file is loaded, and it refers to the index column that was written into the file. This can happen when a DataFrame has an explicit index of a row of numbers or labels, but the file was saved with that index as an additional column, which is then labeled as Unnamed: 0. 

tldr - We can delete the Unnamed: 0 column without losing any essential information because it's just an extra index column that was carried over from the original DataFrame's index.

#### Can we reliably infer missing `cityId` values using `city`?

To answer such question, we need to understand a few things.

#### 25. Do all cities have unique `cityId`? Find cities wuth multiple IDs if any.

```python
city_id_counts = df.groupby('city')['cityId'].nunique()
cities_with_multiple_ids = city_id_counts[city_id_counts > 1]
print("Cities with multiple cityIds:")
print(cities_with_multiple_ids)
```

#### 26. Does every city have at least one `cityId`? Identify any cities without a `cityId` if they exist.

```python
cities_without_city_id = df[df['cityId'].isnull()]['city'].unique()
print("Cities without a cityId:")
print(cities_without_city_id)
```

#### 27. Replace `NaN` values in the `cityId` column with the corresponding `cityId` for cities that have a unique `cityId`. Report how many missing values where in the `cityId` column before and after the replacement.

```python
missing_before = df['cityId'].isnull().sum()

for city in df['city'].unique():
    city_ids = df[df['city'] == city]['cityId'].dropna().unique()
    if len(city_ids) == 1:
        city_id = city_ids[0]
        df.loc[(df['city'] == city) & (df['cityId'].isnull()), 'cityId'] = city_id

missing_after = df['cityId'].isnull().sum()

print(f"Missing values in 'cityId' before replacement: {missing_before}")
print(f"Missing values in 'cityId' after replacement: {missing_after}")
```
