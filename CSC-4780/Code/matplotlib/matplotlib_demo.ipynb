## 1. Intro to Matplotlib


Matplotlib is the foundational plotting library in Python and one of the most prominent data visualization tools. It is highly efficient and capable of producing publication-quality figures in various formats, including PDF, SVG, JPG, PNG, BMP, and GIF. Matplotlib supports a wide range of visualization types, such as line plots, scatter plots, histograms, bar charts, error charts, pie charts, and box plots. It also includes support for 3D plotting.

Many Python libraries, such as pandas and Seaborn, are built on top of Matplotlib, allowing users to leverage its capabilities with less code.

The Matplotlib project was initiated by John Hunter in 2002 to visualize Electrocorticography (ECoG) data from epilepsy patients during his postdoctoral research in neurobiology. Over time, it evolved into the most widely used plotting library in Python. Notably, it was used for data visualization during the landing of the Phoenix spacecraft in 2008.

## 4. Matplotlib Object Hierarchy
Matplotlib organizes plots in a **hierarchical structure**, where a **Figure** contains one or more **Axes**, and each **Axes** contains various components like titles, labels, and plotted data:
- **Figure**: The top-level container that holds everything.
  - **Axes (subplot 1)**: The actual plotting area.
    - **X-axis, Y-axis**: Coordinate system for the plot.
    - **Tick marks**: Define scale on the axes.
    - **Grid lines**: Optional background grid.
    - **Labels and Title**: Text describing the plot.
    - **Lines, markers, bars**: The plotted data.
    - **Legend**: Explains different elements in the plot.
    - **Annotations**: Additional notes or highlights.
  - **Axes (subplot 2)** (Optional second plot inside the same figure)
    - **Same components as Axes 1**.



Why Use `plt.figure()`?
- To explicitly create a new figure (avoids modifying an existing figure).
- When working with multiple figures (to prevent overlapping plots).
- To customize the figure properties, like size or background color.

## 5. Matplotlib Application Programming Interface (API)

Matplotlib provides two primary interfaces for creating visualizations:

### A. Pyplot Interface (State-Based, MATLAB-Style)
   - Designed to mimic MATLABs plotting functions.
   - Simple and convenient for quick plots but less flexible for complex visualizations.
   - Uses a **global state** to track the *current* figure and current axes.
      - All plotting functions modify the current figure such as:
         - Creating a new figure.
         - Adding a plotting area (axes).
         - Plotting data within the current axes.
      - If multiple plots are created, previous figures may be unintentionally modified.
      - Its difficult to track which figure is currently active when working with multiple plots.
         - We can get a reference to the current figure and axes using the following commands:
            - `plt.gcf( )`   # get current figure
            - `plt.gca( )`   # get current axes 

````python
plt.plot([0, 1, 2, 3, 4, 5], [0, 2, 1, 3, 3, 2]);
# Matplotlib automatically creates a figure and axes objects behind the scenes.

# plt.show() # After plt.show(), the figure is no longer active

fig = plt.gcf()
ax = plt.gca()

fig.set_size_inches(7, 2)
ax.set_title("We added a title")
ax.grid(True)
````

#### B. Object-Oriented (OO) Interface
   - Provides **more control and flexibility** over plots.
   - Works by explicitly creating **Figure** and **Axes** objects.
   - Preferred for complex, multi-plot visualizations.

````python
fig, ax = plt.subplots()  
ax.plot([0, 1, 2, 3, 4, 5], [0, 2, 1, 3, 3, 2]);
````

## 7. Object-Oriented Interface in Matplotlib

The Object-Oriented (OO) Interface in Matplotlib provides more control and flexibility for complex visualizations. 



**Why Use the Object-Oriented API?**
- More control over multiple plots in the same figure.  
- Better structure for complex visualizations.  
- No reliance on an "active" state, making it **more explicit and readable**.


| **Command**                     | **Use Case**                                      |
|----------------------------------|--------------------------------------------------|
| `fig, ax = plt.subplots()`       | Creates a **single plot** (one `Axes` object).  |
| `fig, axs = plt.subplots(n, m)`  | Creates **multiple subplots** in a **nm grid** (returns an array of `Axes` objects). |
| `axs[i, j]`                      | Accesses a **specific subplot** within `plt.subplots(rows, cols)`. |


````python
fig, ax = plt.subplots()
ax.hist(np.random.randn(1000));
````

````python
fig, axs = plt.subplots(2, 2)  # Creates a 2x2 grid

axs[0, 0].set_title("1st Subplot")
axs[0, 1].set_title("2nd Subplot")
axs[1, 0].set_title("3rd Subplot")
axs[1, 1].set_title("4th Subplot")


axs[0, 0].hist(np.random.randn(1000)); 
axs[0, 1].hist(1 + 2*np.random.randn(1000)); 
axs[1, 0].hist(3 * np.random.randn(1000)); 
axs[1, 1].hist(-np.random.randn(1000)); 

plt.tight_layout()
````

## 10.	Histogram `.hist()`
A **histogram** is used to visualize the distribution of a dataset by dividing data into **bins** and counting the number of occurrences within each bin. It helps in understanding **frequency distributions, patterns, and outliers** in numerical data.

### Key Features of a Histogram
- Represents **frequency distributions**.
- Groups data into **bins**.
- Helps in analyzing **skewness, spread, and central tendency**.

### Customization Options
| **Parameter** | **Description** | **Example** |
|--------------|----------------|-------------|
| `bins` | Defines the number of bins (intervals) | `bins=20` |
| `color` | Sets the bar colors | `color='blue'` |
| `alpha` | Controls transparency (0-1) | `alpha=0.7` |
| `edgecolor` | Adds a border around bars | `edgecolor='black'` |
| `density` | Normalizes the histogram | `density=True` |
| `cumulative` | Shows cumulative frequency | `cumulative=True` |

### Best Use Cases
- **Understanding distributions** (normal, skewed, uniform).
- **Comparing datasets** (multiple histograms).
- **Detecting outliers and skewness**.



````python
np.random.seed(4)
data = np.random.randn(1000)

fig, ax = plt.subplots(figsize=(7, 5))

ax.hist(data, bins=30, color='blue', edgecolor='black', alpha=0.7)


ax.set_title("Hello, I'm an equal width Histogram!", fontsize=15)
ax.set_xlabel("Value", fontsize=12)
ax.set_ylabel("Frequency", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.5)
````

````python
num_bins = 10 

# Compute bin edges using quantiles
bin_edges = np.quantile(data, np.linspace(0, 1, num_bins + 1))

# Compute frequency of each bin
bin_counts, _ = np.histogram(data, bins=bin_edges)

# Compute bin widths
bin_widths = np.diff(bin_edges)

# Compute density: count / (bin_width * total_count)
bin_densities = bin_counts / (bin_widths * len(data))


fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(bin_edges[:-1], bin_densities, width=bin_widths, align="edge", color="blue", edgecolor="black", alpha=0.7)

# Customize plot
ax.set_title("Hello, I'm an equal frequency Histogram!", fontsize=14)
ax.set_xlabel("Value", fontsize=12)
ax.set_ylabel("Density", fontsize=12)
````

## 12.	Error Bar Chart `.errorbar()`

Measurements often contain **inherent variability** due to imperfections in precision. To account for this, repeated measurements are taken, resulting in a **set of values** rather than a single precise reading.

A **mean value** is typically used to represent the dataset, while an **error bar** visually conveys the **spread or uncertainty** of the data distribution.

Matplotlib provides the `errorbar()` function to effectively illustrate these variations, making it useful for displaying:
- **Measurement uncertainty** in scientific experiments.
- **Confidence intervals** for statistical estimates.
- **Variability in repeated observations**.



````python
x = np.arange(0, 4, 0.2)
y = np.exp(-x)
error = 0.1 * np.abs(np.random.randn(len(y)))

fig, ax = plt.subplots(figsize=(7, 5))
ax.errorbar(x, y, yerr=error, fmt='.-r', capsize=3, capthick=2);
````

## 13. Stacked Bar Chart `.bar(bottom= )`


A **stacked bar chart** visualizes **part-to-whole relationships** by stacking multiple data series **on top of each other** within a single bar. 

````python
data_a = np.array([15., 30., 45., 20.])
data_b = np.array([5., 25., 10., 20.])
categories = ["Cat 1", "Cat 2", "Cat 3", "Cat 4"]

# Create figure and axis
fig, ax = plt.subplots(figsize=(7, 5))

# Plot stacked bars
ax.bar(categories, data_a, color="blue", edgecolor="black", alpha=0.7, label="Group A")
ax.bar(categories, data_b, bottom=data_a, color="red", edgecolor="black", alpha=0.7, label="Group B")
ax.legend();
````

The **`bottom`** parameter in `plt.bar()` defines the **starting position** of each bar. By default, bars start from **zero** and extend to their respective values. However, when `bottom` is specified, the bars **stack on top of the given values** instead of starting from zero.



## 14. Pie Chart `.pie()`


A **pie chart** is a circular representation of data, divided into **sectors (wedges)**. Each wedge's **arc length and area** are proportional to the corresponding value in the dataset.

#### When to Use Pie Charts?
- Best suited for **comparing a part to the whole**, rather than comparing individual parts.
- Useful for **percentage-based data** or **relative proportions**.
- Not ideal for datasets with **too many categories** or **similar values** (bar charts may be better).

#### How Pie Charts Work in Matplotlib
Matplotlib provides the **`pie()`** function to create pie charts. Given an array `X`:
- Each value in `X` represents a **wedge**.
- Wedges are proportional to **`x / sum(X)`**, ensuring they sum to 100%.
- Additional options allow for **labels, colors, exploding wedges, and percentage formatting**.

````python
x = [35, 25, 20, 20]
labels = ['Computer Science', 'Electronics', 'Mechanical', 'Chemistry']

# Create figure with a proper aspect ratio
fig, ax = plt.subplots(figsize=(5, 5))

# Plot pie chart with enhancements
ax.pie(
    x, 
    labels=labels, 
    autopct='%1.1f%%',  # Display percentage values
    startangle=90,  # Rotate chart for better readability
    colors=['red', '#6DAEDB', '#89CFF0', '#A7D8F3'],  # Custom colors
    wedgeprops={'edgecolor': 'black'}  # Add border for clarity
);
````

## 15. Boxplot `boxplot()`

A **boxplot** (or **box-and-whisker plot**) is a visualization tool used to compare distributions of values by summarizing key statistical measures, including the **median, quartiles, minimum, and maximum** of a dataset.

#### How Boxplots Work
The **`boxplot()`** function in Matplotlib computes the following statistical quantities:

- **Median (Red Bar)**: The central value of the dataset, dividing it into two equal halves.
- **Interquartile Range (IQR) (Blue Box)**: The range between the **lower quartile (Q1, 25th percentile)** and **upper quartile (Q3, 75th percentile)**, covering **50% of the data**.
- **Whiskers**:
  - The **lower whisker** extends to the **lowest value** within **1.5IQR** from Q1.
  - The **upper whisker** extends to the **highest value** within **1.5IQR** from Q3.
- **Outliers (Cross Markers)**: Values beyond **1.5IQR** from the quartiles are considered outliers and displayed as separate points.

#### Why Use Boxplots?
- **Compare multiple distributions** efficiently.
- **Identify outliers** and variability in datasets.
- **Understand symmetry and skewness** of the data.

````python
data = np.random.randn(100)

fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot(data);
````

````python
data1 = np.random.randn(100)
data2 = np.random.randn(100) + 1

fig, ax = plt.subplots(figsize=(7, 5))
ax.boxplot([data1, data2], widths=0.4)
ax.set_xticklabels(["Dataset 1", "Dataset 2"], fontsize=12);
````

## 19. Handling axes


Matplotlib automatically sets the limits of the plot to precisely contain the plotted datasets. Sometimes, we want to set the
axes limits ourself. We can set the axes limits with the **axis()** function as follows:

````python
x = np.arange(1, 5)

plt.plot(x, x*1.5, x, x*3.0, x, x/3.0)

plt.axis()   # shows the current axis limits values

plt.axis([0, 5, -1, 13]);
````

We can see that we now have more space around the lines.

If we execute **axis()** without parameters, it returns the actual axis limits.

We can set parameters to **axis()** by a list of four values. 

The list of four values are the keyword arguments [xmin, xmax, ymin, ymax] allows the minimum and maximum limits for X and Y 
axis respectively.
 

We can control the limits for each axis separately using the `xlim()` and `ylim()` functions. This can be done as follows:-

````python
x = np.arange(1, 5)

plt.plot(x, x*1.5, x, x*3.0, x, x/3.0)

plt.xlim([1.0, 4.0])
plt.ylim([0.0, 12.0]);
````

Legends for curves in a figure can be added in two ways. One method is to use the **legend** method of the axis object and 
pass a list/tuple of legend texts as follows:-

````python
x = np.arange(1, 5)

fig, ax = plt.subplots()
ax.plot(x, x*1.5)
ax.plot(x, x*3.0)
ax.plot(x, x/3.0)

ax.legend(['Normal','Fast','Slow']);
````

The above method follows the MATLAB API. It is prone to errors and unflexible if curves are added to or removed from the plot.
It resulted in a wrongly labelled curve.

A better method is to use the **label** keyword argument when plots are added to the figure. Then we use the **legend** method without arguments to add the legend to the figure. 

The advantage of this method is that if curves are added or removed from the figure, the legend is automatically updated
accordingly. It can be achieved by executing the code below:-


````python
x = np.arange(1, 5)

fig, ax = plt.subplots()
ax.plot(x, x*1.5, label='Normal')
ax.plot(x, x*3.0, label='Fast')
ax.plot(x, x/3.0, label='Slow')

ax.legend();
````

#### Customizing Legend Location in Matplotlib

The `legend()` function in Matplotlib allows you to add a legend to a plot and control its position within the figure using the `loc` parameter. 

You can specify the location using either an **integer code** or a **descriptive string**.

**Legend Positioning Options**

| Integer Code | String Equivalent | Legend Position        |
|-----------------|----------------------|---------------------------|
| `0`            | `"best"`              | Automatic placement (Matplotlib decides) |
| `1`            | `"upper right"`       | Upper right corner |
| `2`            | `"upper left"`        | Upper left corner |
| `3`            | `"lower left"`        | Lower left corner |
| `4`            | `"lower right"`       | Lower right corner |
| `5`            | `"right"`             | Right side of the plot |
| `6`            | `"center left"`       | Center left |
| `7`            | `"center right"`      | Center right |
| `8`            | `"lower center"`      | Lower center |
| `9`            | `"upper center"`      | Upper center |
| `10`           | `"center"`            | Center of the plot |





**Placing the Legend Outside the Axes**

In many cases, you may want to position the legend outside the plot area to avoid overlap with the data. 


- The `bbox_to_anchor` parameter in `plt.legend()` allows you to move the legend outside the axes. It specifies a bounding box where the legend is positioned, with `loc` determining which part of the legend box aligns with the specified coordinates. 
    - For example, setting `bbox_to_anchor=(1.05, 1)` places the upper-left corner of the legend at `(1.05, 1)`, slightly outside the plot area. 
   
- Additionally, `frameon=False` can be used to remove the border around the legend.

````python
x = np.arange(1, 5)

fig, ax = plt.subplots()
ax.plot(x, x*1.5, label='Normal')
ax.plot(x, x*3.0, label='Fast')
ax.plot(x, x/3.0, label='Slow')

ax.legend(bbox_to_anchor=(1.05,1), loc='upper left', frameon=False);
````

## Specifying Colors in Matplotlib

Matplotlib allows colors to be specified in multiple ways, including **abbreviations, full names, hex codes, RGB tuples, and grayscale values**.

#### Color Abbreviations and Names
The table below lists the available **color abbreviations** and their corresponding **full names**:

| **Color Abbreviation** | **Color Name** |
|------------------------|---------------|
| `b` | Blue |
| `c` | Cyan |
| `g` | Green |
| `k` | Black |
| `m` | Magenta |
| `r` | Red |
| `w` | White |
| `y` | Yellow |

---

#### Other Ways to Specify Colors
In addition to abbreviations, colors can be specified using:

**Full Color Name**  
   - Example: `"yellow"`, `"red"`, `"cyan"`

**Hexadecimal String** (Similar to HTML/CSS colors)  
   - Example: `"#FF00FF"` (Magenta), `"#00FF00"` (Green)

**RGB Tuples** (Red, Green, Blue values ranging from `0` to `1`)  
   - Example: `(1, 0, 1)` for Magenta

**Grayscale Intensity** (A string representing a shade of gray)  
   - Example: `"0.7"` (A light gray, where `"0"` is black and `"1"` is white)




````python
x = np.linspace(0, 10, 100)

# Create a figure with subplots (3 rows, 2 columns)
fig, axs = plt.subplots(3, 2, figsize=(10, 8))

# Remove empty subplot (since we only have 5 plots)
fig.delaxes(axs[2, 1])


axs[0, 0].plot(x, np.sin(x), 'r')
axs[0, 0].set_title("Red (Abbreviation)")
axs[0, 0].axis('off')

axs[0, 1].plot(x, np.cos(x), color='cyan')
axs[0, 1].set_title("Cyan (Full Name)")
axs[0, 1].axis('off')

axs[1, 0].plot(x, -np.sin(x), color='#FF00FF')
axs[1, 0].set_title("Magenta (Hex)")
axs[1, 0].axis('off')

axs[1, 1].plot(x, -np.cos(x), color=(0, 1, 0))
axs[1, 1].set_title("Green (RGB Tuple)")
axs[1, 1].axis('off')

axs[2, 0].plot(x, np.sin(x)/2, color='0.5')
axs[2, 0].set_title("Gray (Grayscale)")
axs[2, 0].axis('off')

plt.tight_layout()
plt.suptitle("Different Ways to Specify Colors in Matplotlib", fontsize=14,  y=1.06);
````
